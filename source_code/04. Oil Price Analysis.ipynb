## 4. 데이터 정리

```python
import pandas as pd
from glob import glob
```

```python
# 파일 목록 한번에 가져오기
glob("../data/지역_*.xls")
```

```python
# 파일명 저장
stations_files = glob("../data/지역_*.xls")
stations_files[:5]
```

```python
# 하나만 읽어보기
tmp = pd.read_excel(stations_files[0], header = 2)
tmp.tail(2)
```

```python
tmp_raw = []
for file_name in stations_files:
    tmp = pd.read_excel(file_name, header = 2)
    tmp_raw.append(tmp)
```

```python
stations_raw = pd.concat(tmp_raw)
stations_raw
```

```python
stations_raw.info()
```

```python
stations = pd.DataFrame({
    "상호" : stations_raw["상호"],
    "주소" : stations_raw["주소"],
    "가격" : stations_raw["휘발유"],
    "셀프" : stations_raw["셀프여부"],
    "상표" : stations_raw["상표"]
})
stations.head()
```

```python
for eachAddress in stations["주소"]:
    print(eachAddress.split()[1])
```

```python
stations["구"] = [eachAddress.split()[1] for eachAddress in stations["주소"]]
```

```python
stations
```

```python
stations["구"].unique(), len(stations["구"].unique())
```

```python
# 가격 데이터형 변환 object => float
stations["가격"] = stations["가격"].astype(float)
```

```python
# 가격 정보가 없는 주유소
stations[stations["가격"] == "-"]
```

```python
# 가격 정보가 있는 주유소만 사용
stations = stations[stations["가격"] != "-"]
stations.tail()
```

```python
stations["가격"] = stations["가격"].astype(float)
```

```python
stations.info()
```

```python
stations.reset_index(inplace = True)
```

```python
stations
```

```python
del stations["index"]
stations
```

```python
# 가격별 주유소 정렬
stations.sort_values(by = "가격", ascending=True).head()
```

```python
# 구별 평균 가격 확인
import numpy as np
gu_data = pd.pivot_table(stations, index = ["구"], values = "가격", aggfunc=np.mean)
```
