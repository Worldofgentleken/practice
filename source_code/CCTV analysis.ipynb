```python
# 원하는 모듈이 설치되어 있다면 import 명령을 통해 사용하겠다고 선언한다.
import pandas as pd
```

- 통상 csv는 띄어쓰기로 구분되니 그냥 read_csv 명령으로 읽기만 하면 된다.
- 긴 파일명을 끝까지 입력하지 말고 적당한 곳에서 TAB 키를 누르면 자동완선
- 한글은 encoding 설정이 필수

```python
CCTV_Seoul = pd.read_csv('../data/01. Seoul_CCTV.csv', encoding = 'utf-8')
CCTV_Seoul.head()
```

## 컬럼 이름 조회
- 변수명.columns : column의 이름만 조회하고 싶을때 사용
- 변수명.columns[i] : index 번호를 대괄호안에 넣으면 해당 컬럼의 이름만 조회가능

```python
CCTV_Seoul.columns[0]
```

## 컬럼 이름 변경
### rename() 함수 사용
- 변수명.rename(columns = {변수명.columns[인덱스번호]: '바꿀 컬럼명'}, inplace = True)
- inplace = True를 해줘야 데이터가 업데이트 된다.

```python
CCTV_Seoul.rename(columns = {CCTV_Seoul.columns[1] : '구별'}, inplace= True)
CCTV_Seoul.head()
```

```python
CCTV_Seoul.rename(columns={CCTV_Seoul.columns[2]:'총계'},inplace=True)
CCTV_Seoul.head()
```

```python
CCTV_Seoul = CCTV_Seoul.drop(['순번'], axis = 1)
CCTV_Seoul.head()
```

```python
CCTV_Seoul.rename(columns={CCTV_Seoul.columns[2]:'2015년 이전 설치'},inplace=True)
CCTV_Seoul.head()
```

## Excel 파일 읽기
- 변수명 = pd.read_excel('파일경로')

```python
pop_Seoul = pd.read_excel('../data/01. Seoul_Population.xlsx')
pop_Seoul.head()
```

## Excel 파일 정제
- 엑셀파일 불러오기
- 자료를 읽기 시작하는 행(header)를 지정
- 읽어올 엑셀의 컬럼을 지정(usecols)

```python
pop_Seoul = pd.read_excel('../data/01. Seoul_Population.xlsx', header = 2, usecols = 'B, D, G, J, N')
pop_Seoul.head()
```

## 컬럼의 이름 바꾸기

```python
pop_Seoul.rename(columns = {
    pop_Seoul.columns[0] : '구별',
    pop_Seoul.columns[1] : '인구수',
    pop_Seoul.columns[2] : '한국인',
    pop_Seoul.columns[3] : '외국인',
    pop_Seoul.columns[4] : '고령자'
}, inplace = True)
pop_Seoul.head()
```

## CCTV 데이터 훑어보기

```python
CCTV_Seoul.info()
```

```python
CCTV_Seoul.loc[:, '총계' : '2023년'] = CCTV_Seoul.loc[:, '총계' : '2023년'].applymap(lambda x: int(x.replace(',', '')))
CCTV_Seoul.head()
```

```python
CCTV_Seoul.sort_values(by = '총계', ascending= False).head()
```

```python
CCTV_Seoul.sort_values(by = '총계').head()
```

```python
print(CCTV_Seoul['2015년 이전 설치'].unique())
```

```python
CCTV_Seoul = CCTV_Seoul[CCTV_Seoul['2015년 이전 설치'] != 0]
CCTV_Seoul['최근증가율'] = (
    (CCTV_Seoul['2015년'] + CCTV_Seoul['2016년'] + CCTV_Seoul['2017년'] + CCTV_Seoul['2018년'] + CCTV_Seoul['2019년']
     + CCTV_Seoul['2020년'] + CCTV_Seoul['2021년'] + CCTV_Seoul['2022년'] + CCTV_Seoul['2023년']) / CCTV_Seoul['2015년 이전 설치']*100)
```

```python
CCTV_Seoul.sort_values(by = '최근증가율', ascending= False)
```

```python
CCTV_Seoul = CCTV_Seoul.drop([0])
```

```python
CCTV_Seoul
```

## 인구 현황 데이터 훑어보기

```python
pop_Seoul.head()
```

```python
pop_Seoul = pop_Seoul.drop([0])
```

```python
pop_Seoul
```

```python
pop_Seoul['외국인비율'] = pop_Seoul['외국인'] / pop_Seoul['인구수'] * 100
pop_Seoul['고령자비율'] = pop_Seoul['고령자'] / pop_Seoul['인구수'] * 100
```

```python
pop_Seoul.head()
```

```python
data_result = pd.merge(CCTV_Seoul, pop_Seoul, on = '구별')
data_result.head()
```

## 년도별 데이터 컬럼 삭제
- del : 한 개의 컬럼을 삭제할 때 용이
- drop() : 여러개의 row나 columns을 삭제할 때 용이

```python
del data_result['2015년 이전 설치']
```

```python
data_result = data_result.drop(['2015년','2016년', '2017년', '2018년','2019년', '2020년', '2021년'],  axis = 1)
```

```python
data_result.head()
```

```python
del data_result['2022년']
del data_result['2023년']
```

```python
data_result.head()
```

## 인덱스 변경
- set_index()
- 선택한 컬럼을 데이터 프레임의 인덱스로 지정

```python
data_result.set_index('구별', inplace = True)
data_result.head()
```
