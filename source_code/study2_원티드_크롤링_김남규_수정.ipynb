```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
from bs4 import BeautifulSoup
import pandas as pd 
from urllib.parse import urljoin
```

```python
driver = webdriver.Chrome("")
time.sleep(3)
url_base='https://www.wanted.co.kr'
url_sub='/wdlist/518/1634?country=kr&job_sort=job.recommend_order&years=-1&selected=1634&locations=seoul.songpa-gu&locations=seoul.gangnam-gu&locations=seoul.guro-gu&locations=seoul.geumcheon-gu&locations=seoul.seocho-gu'
url=url_base + url_sub
time.sleep(3)
driver.get(url)
time.sleep(3)
```

```python
initial_scroll = 300  # 처음에 스크롤할 양
scroll_step = 100  # 스크롤 스텝
scroll_pause_time = 0.5  # 스크롤 후 대기 시간

company_list = []
title_list = []
url_add = []


# 페이지 처음에 일부 스크롤
driver.execute_script(f'window.scrollBy(0, {initial_scroll});')
time.sleep(scroll_pause_time)

last_height = driver.execute_script('return document.body.scrollHeight')

while True:
    req = driver.page_source
    soup = BeautifulSoup(req, 'html.parser')
    soup_list = soup.find_all('li', class_='Card_Card__hfQPt')

    for item in soup_list:
        company_element = item.find('p', class_ = 'Typography_Typography__root__xYuMs Typography_Typography__label2__BRssq Typography_Typography__weightMedium__O0Rdi JobCard_JobCard__body__company__F6XoH')
        company = company_element.text if company_element else None

        #title_element = item.find('p',class_='Typography_Typography__root__xYuMs Typography_Typography__body2__EpxWz Typography_Typography__weightBold__e15ql JobCard_JobCard__body__position__P8R0W')
        #title = title_element.text if title_element else None ~~~오류가 난다 빈문장 출력
        time.sleep(3)
        title_element =item.find('p', 'Typography_Typography__root__xYuMs Typography_Typography__body1__UmyQW Typography_Typography__weightBold__e15ql JobCard_JobCard__body__position__P8R0W').text
#         title=title_element.text if title_element else None
        time.sleep(3)
        
        
        job_url_element = item.find('a')
        job_url = urljoin(url_base, job_url_element['href']) if job_url_element else None

        if job_url and job_url not in url_add:
            company_list.append(company)
            title_list.append(title_element)
            url_add.append(job_url)

    driver.execute_script(f'window.scrollBy(0, {last_height});')
    time.sleep(scroll_pause_time)
    new_height = driver.execute_script('return document.body.scrollHeight')
    if new_height == last_height:
        break
    last_height = new_height

print("수집된 데이터 개수:", len(company_list), len(title_list), len(url_add))

# 드라이버 종료
#driver.quit()
```
