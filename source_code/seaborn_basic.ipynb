# Seaborn

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import rc

plt.rcParams['axes.unicode_minus'] = False
rc('font', family = 'Arial Unicode MS')
%matplotlib inline
```

#### 예제 1 : seaborn 기초

```python
np.linspace(0, 14, 100)
```

```python
x = np.linspace(0, 14, 100)
y1 = np.sin(x)
y2 = 2 * np.sin(x + 0.5)
y3 = 3 * np.sin(x + 1.0)
y4 = 4 * np.sin(x + 1.5)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x, y1, x, y2, x, y3, x, y4)
plt.show()
```

```python
# sns.set_style()
# 'white', 'whitegrid', 'dark', 'darkgrid'
sns.set_style('white')
plt.figure(figsize = (10, 6))
plt.plot(x, y1, x, y2, x, y3, x, y4)
plt.show()
```

```python
sns.set_style('dark')
plt.figure(figsize = (10, 6))
plt.plot(x, y1, x, y2, x, y3, x, y4)
plt.show()
```

```python
sns.set_style('whitegrid')
plt.figure(figsize = (10, 6))
plt.plot(x, y1, x, y2, x, y3, x, y4)
plt.show()
```

```python
sns.set_style('darkgrid')
plt.figure(figsize = (10, 6))
plt.plot(x, y1, x, y2, x, y3, x, y4)
plt.show()
```

#### 예제2 : seaborn tips data
- boxplot
- swarmplot
- lmplot

```python
tips = sns.load_dataset('tips')
tips
```

```python
tips.info()
```

```python
# boxplot
plt.figure(figsize= (8, 6))
sns.boxplot(x = tips['total_bill'])
plt.show()
```

```python
plt.figure(figsize= (8, 6))
sns.boxplot(x = 'day', y = 'total_bill', data = tips)
plt.show()
```

```python
tips.head()
```

```python
# boxplot hue, palette option
# hue : 카테고리 데이터를 표현하는 옵션
plt.figure(figsize=(8, 6))
sns.boxplot(x = 'day', y = 'total_bill', data = tips, hue = 'smoker', palette = 'Set3') # Set 1부터 3까지 있은ㅁ
plt.show()
```

```python
# swarmplot
# color : 0 부터 1사이 검은색 부터 흰색 사이 값을 조절
plt.figure(figsize=(8, 6))
sns.swarmplot(x = 'day', y = 'total_bill', data = tips, color = '0.5')
plt.show()
```

```python
# boxplot with swarmplot
plt.figure(figsize=(8, 6))
sns.boxplot(x = 'day', y = 'total_bill', data = tips)
sns.swarmplot(x = 'day', y = 'total_bill', data = tips, color = '0.2')
plt.show()
```

```python
# lmplot : total_bill과 tip 사이의 관계 파악

sns.set_style('darkgrid')
sns.lmplot(x = 'total_bill', y = 'tip', data = tips, height = 7)
plt.show()
```

```python
# hue optioin

sns.set_style('darkgrid')
sns.lmplot(x = 'total_bill', y = 'tip', data = tips, height = 7, hue = 'smoker')
plt.show()
```

#### 예제3 : flight data
- heatmap

```python
flights = sns.load_dataset('flights')
flights
```

```python
flights.info()
```

```python
# pivot
# index, columns, values, aggfunc
flights = flights.pivot(index = 'month', columns = 'year', values = 'passengers')
flights.head()
```

```python
# heatmap

plt.figure(figsize=(10, 8))
sns.heatmap(data = flights, annot = True, fmt = 'd') # annot = True 데이터 값 표시, fmt = 'd' 정수형 표시
plt.show()
```

```python
# colormap
plt.figure(figsize=(10, 8))
sns.heatmap(flights, annot = True, fmt = 'd', cmap = 'YlGnBu')
plt.show()
```

#### 예제4 : iris data
- pairplot

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris.info()
```

```python
# pairplot
sns.set_style('ticks')
sns.pairplot(iris)
plt.show()
```

```python
iris.head()
```

```python
iris['species'].unique()
```

```python
# hue option

sns.pairplot(iris, hue = 'species')
plt.show()
```

```python
# 원하는 컬럼만 pairplot

sns.pairplot(iris, hue = 'species', x_vars=['sepal_width', 'sepal_length'], y_vars=['petal_width', 'petal_length'])
plt.show()
```

#### 예제5 : anscombe data
- lmplot

```python
anscombe = sns.load_dataset('anscombe')
anscombe.head()
```

```python
anscombe.info()
```

```python
anscombe['dataset'].unique()
```

```python
sns.set_style('darkgrid')
sns.lmplot(x = 'x', y = 'y', data = anscombe.query("dataset == 'I'"), ci = None, height = 7)
plt.show()
```

```python
sns.set_style('darkgrid')
sns.lmplot(x = 'x', y = 'y', data = anscombe.query("dataset == 'I'"), ci = None, height = 7, scatter_kws={'s' : 80})
plt.show()
```

```python
# order option
sns.set_style('darkgrid')
sns.lmplot(x = 'x',
           y = 'y',
           data = anscombe.query("dataset == 'II'"),
           order = 1,
           ci = None,
           height = 7,
           scatter_kws={'s' : 80})
plt.show()
```

```python
# order option
sns.set_style('darkgrid')
sns.lmplot(x = 'x',
           y = 'y',
           data = anscombe.query("dataset == 'II'"),
           order = 2,
           ci = None,
           height = 7,
           scatter_kws={'s' : 80})
plt.show()
```

```python
# outlier
sns.set_style('darkgrid')
sns.lmplot(x = 'x',
           y = 'y',
           data = anscombe.query("dataset == 'III'"),
           ci = None,
           height = 7,
           scatter_kws={'s' : 80})
plt.show()
```
