### 삼각함수 그리기
- np.arange(a, b, s) : a부터 b까지 s의 간격
- np.sin(value)

```python
import numpy as np

t = np.arange(0, 12, 0.01)
y = np.sin(t)
```

```python
plt.figure(figsize = (10, 6))
plt.plot(t, np.sin(t))
plt.plot(t, np.cos(t))
plt.show()
```

- 1. 격자무늬 추가
- 2. 그래프 제목 추가
- 3. x축, y축 제목 추가
- 4. 주황색, 파란색 선 데이터 의미 구분

```python
def drawGraph():

    plt.figure(figsize = (10, 6))
    plt.plot(t, np.sin(t), label = 'sin')
    plt.plot(t, np.cos(t), label = 'cos')
    plt.grid(True)
    plt.legend(loc = 'upper right')
    plt.title('Example of sinwave')
    plt.xlabel('time')
    plt.ylabel('Amplitude')
    plt.show()
```

```python
drawGraph()
```

### 그래프 커스텀

```python
# t = [0, 1, 2, 3, 4, 5, 6]
t = list(range(0, 7))
y = [1, 4, 5, 8, 9, 5, 3]
```

```python
def drawGraph():
           
    plt.figure(figsize = (10, 6))
    plt.plot(
        t,
        y,
        color = 'green',
        linestyle = 'dashed',
        marker = 'o',
        markerfacecolor = 'blue',
        markersize = 15
    )
    
    plt.xlim([-0.5, 6.5])
    plt.ylim([0.5, 9.5])
    plt.show()

drawGraph()
```

### satter plot

```python
t = np.array(range(0,10))
y = np.array([9, 8, 7, 9, 8, 3, 2, 4, 3, 4])
```

```python
def drawGraph():
        
    plt.figure(figsize=(10,6))
    plt.scatter(t, y)
    plt.show

drawGraph()
```

```python
colormap = t

def drawGraph():
        
    plt.figure(figsize=(10,6))
    plt.scatter(t, y, s = 50, c = colormap, marker = '>')
    plt.colorbar()
    plt.show

drawGraph()
```
